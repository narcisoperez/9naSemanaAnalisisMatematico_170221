## Qu es una ecuacin en derivadas parciales?

Una [Ecuacin en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) es una ecuacin que, como su nombre lo indica, que contiene [derivadas parciales](https://es.wikipedia.org/wiki/Derivada_parcial). A diferencia de lo que habamos visto con las [ecuaciones diferenciales ordinarias](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/), en donde la funcin incgnita depende **solo** de *una variable*; en las [Ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales), o [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) para abreviar, la funcin incgnita va a depender de **dos o ms** *variables independientes* $x, y, \dots$. Generalmente a la funcin incgnita la vamos a expresar como $u(x, y, \dots)$ y a sus derivadas parciales como $\partial u / \partial x = u_x$ o  $\partial u / \partial y = u_y$ dependiendo de sobre que variable estemos derivando. Entonces una [Ecuacin en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) va a ser la identidad que relaciona a las *variables independientes* ($x, y, \dots$), con la *variable dependiente* $u$ (nuestra funcin incgnita), y las [derivadas parciales](https://es.wikipedia.org/wiki/Derivada_parcial) de $u$. Lo podramos expresar de la siguiente forma:

$$F(x, y, u(x, y), u_x(x, y), u_y(x, y)) = F(x, y, u, u_x, u_y) = 0$$

Al igual de como pasaba con las [ecuaciones diferenciales ordinarias](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/), el ***orden*** de una [Ecuacin en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) va a estar dado por la mayor [derivada](https://es.wikipedia.org/wiki/Derivada) presente. Por lo tanto, el caso que expresamos ms arriba corresponde a una [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales), con dos *variables independientes*, de *primer orden*. Si quisiramos expresar una [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de *segundo orden*, podramos hacerlo de la siguiente manera:

$$F(x, y, u, u_x, u_y, u_{xx}, u_{xy}, u_{yy})=0.$$

Linear regression is a technique used in modeling the linear relationship between an input and its output. Given a set of known input/output values, linear regression will find the linear function that best fits these values and that can predict the outcome output for any new input value.

For multiple inputs, the process is called Multiple Linear Regression.

### What problem does it solve?
Imagine you know multiple inputs and one output for a problem, and you want to be able to compute the output for any new input.

For instance, lets say you know the characteristics of houses (size, age, number of rooms, etc.) and the selling prices in Amsterdam for the last month, and you would like to predict the selling price of a new house, knowing its size.

Linear regression will find the best prediction given our example training set.

Single input-output problem
For clarity, we will restrain our model to one input `x` and one output `y`. This will make it easier to plot the data and follow the graph alongside the explanation. For our example, we will predict the price of a house based on its size.

### Clasificacin de ecuaciones en derivadas parciales

La clasificacin de las [Ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) va a ser algo fundamental, ya que la teora y los mtodos para poder solucionarlas van a depender de la *clase* de ecuacin con la que estemos tratando. Las clasificaciones ms importantes que debemos tener en cuenta, son:

1- **El *orden* de la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales):** Como ya mencionamos, el mismo va a estar dado por el *orden* de la mayor [derivada](https://es.wikipedia.org/wiki/Derivada) presente.


2- **Nmero de variables:** Esta clasificacin va a estar dada por la cantidad de *variables independientes* que contenga la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales).


3- **Linealidad:** Esta es una de las clasificaciones ms importantes, vamos a poder clasificar a las [Ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) en [lineales](https://es.wikipedia.org/wiki/Lineal) o [no lineales](https://es.wikipedia.org/wiki/No_linealidad). En las [lineales](https://es.wikipedia.org/wiki/Lineal), la *variable dependiente* $u$ y todas sus derivadas, van a aparecer en una forma *lineal*, es decir, que van a tener *grado* uno (no van a estar elevadas al cuadrado, o multiplicadas entre s). Ms precisamente, una [Ecuacin en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de segundo orden con dos variables, va a tomar la siguiente forma:

$$Au_{xx} + Bu_{xy} + Cu_{yy} + Du_x + Eu_y + Fu = G$$

en donde $A, B, C, D, E, F,$ y $G$ pueden ser *<a href="https://es.wikipedia.org/wiki/Constante_(matem%C3%A1ticas)">constantes</a>* o una *funcin* dada de $x$ e $y$. Por ejemplo:

$u_{tt} = e^tu_{xx} + \sin t$,    sera una ecuacin [lineal](https://es.wikipedia.org/wiki/Lineal).

$uu_{xx} + u_y = 0$,  sera una ecuacin [no lineal](https://es.wikipedia.org/wiki/No_linealidad).

$u_{xx} + yu_{yy} + u_x = 0$, sera una ecuacin [lineal](https://es.wikipedia.org/wiki/Lineal).

$xu_x + yu_y + u^2 = 0$, sera una ecuacin [no lineal](https://es.wikipedia.org/wiki/No_linealidad).

**Tipos de ecuaciones lineales:** Asimismo, a las [Ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) [lineales](https://es.wikipedia.org/wiki/Lineal) de segundo orden las vamos a poder subdividir en las siguientes categoras:

* **[Ecuaciones parablicas](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_parab%C3%B3lica_en_derivadas_parciales)**: Las cuales van a describir el *[flujo del calor](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_del_calor)* y el proceso de *[difusin](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_de_difusi%C3%B3n)*. stas ecuaciones van a satisfacer la condicin $B^2 -4AC = 0$.

* **[Ecuaciones hiperblicas](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_hiperb%C3%B3lica_en_derivadas_parciales)**: Las cuales describen los *[sistemas de vibracin](https://es.wikipedia.org/wiki/Vibraci%C3%B3n)* y los *[movimientos de ondas](https://es.wikipedia.org/wiki/Onda)*. Satisfacen la condicin $B^2 -4AC > 0$.

* **[Ecuaciones Elpticas](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_el%C3%ADptica_en_derivadas_parciales)**: Las cuales describen los fenmenos de *[estados estacionarios](https://es.wikipedia.org/wiki/Estado_estacionario)* y satisfacen la condicin $B^2 -4AC < 0$.

4- **Homogeneidad:** Otra clasificacin que podemos utilizar, es la de homogeneidad. Una ecuacin va a ser *homognea*, si el lado derecho de la ecuacin, $G(x, y)$, es idnticamente cero para todo $x$ e $y$. En caso contrario, se la llama *no homognea*.

5- **Tipos de coeficientes:** Por ltimo, podemos clasificar a las [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de acuerdo a sus coeficientes $A, B, C, D, E, $ y $F$, si los mismos son *<a href="https://es.wikipedia.org/wiki/Constante_(matem%C3%A1ticas)">constantes</a>*, se dice que la ecuacin es de *coeficientes constantes*, en caso contrario ser de *coeficientes variables*.

### Cmo resolver ecuaciones en derivadas parciales?

Existen varios mtodos que podemos utilizar para intentar resolver las [Ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales), la principal idea detrs la mayora de estos mtodos es la transformar a estas ecuaciones en [ecuaciones diferenciales ordinarias](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/) ([EDO](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_diferencial_ordinaria)), o en alguna [ecuacin algebraica](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_algebraica); las cuales son ms sencillas de resolver. Algunos los mtodos que podemos utilizar son:

1- **El [mtodo de separacin de variables](https://es.wikipedia.org/wiki/M%C3%A9todo_de_separaci%C3%B3n_de_variables):** Este mtodo es uno de los ms importantes y ms productivos a la hora de resolver a las [Ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales). La idea es reducir a la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de $n$ variables, en $n$ [EDOs](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_diferencial_ordinaria).

2- **El mtodo de la [transformada integral](https://es.wikipedia.org/wiki/Transformada_integral):** Este mtodo es similar al que ya vimos al resolver [EDOs](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/). La idea es aplicar una [transformada integral](https://es.wikipedia.org/wiki/Transformada_integral) para reducir una [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de $n$ variables, en otra de $n - 1$ variables. De esta forma una [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de 2 variables, puede ser transformada en una [EDO](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_diferencial_ordinaria).

3- **El mtodo del [cambio de coordenadas](https://es.wikipedia.org/wiki/Sistema_de_coordenadas#Cambios_de_coordenadas):** Este mtodo intenta cambiar a la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) original en una [EDO](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_diferencial_ordinaria) o en otra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) ms sencilla de resolver por medio del cambio de coordenadas del problema.

4- **El mtodo de [perturbacin](https://es.wikipedia.org/wiki/Teor%C3%ADa_perturbacional):** Este mtodo aplica la [teora perturbacional](https://es.wikipedia.org/wiki/Teor%C3%ADa_perturbacional) para intentar cambiar un problema de [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) [no lineal](https://es.wikipedia.org/wiki/No_linealidad) en una serie de problemas [lineales](https://es.wikipedia.org/wiki/Lineal) que se aproximan al [no lineal](https://es.wikipedia.org/wiki/No_linealidad) original.

5- **El mtodo de expansin de [autofunciones](https://es.wikipedia.org/wiki/Autofunci%C3%B3n):** Este mtodo intentan encontrar la solucin de una [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) como una suma infinita de [autofunciones](https://es.wikipedia.org/wiki/Autofunci%C3%B3n). Estas [autofunciones](https://es.wikipedia.org/wiki/Autofunci%C3%B3n) son halladas por medio de la resolucin del problema del [valor propio](https://es.wikipedia.org/wiki/Vector_propio_y_valor_propio) que corresponde al problema original.

6- **El mtodo de las [ecuaciones integrales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_integral)**: Este mtodo convierte a la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) en una [ecuacin integral](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_integral); la cual luego es resuelta aplicando ciertas tcnicas particulares que se aplican a ese tipo ecuaciones.

7- **Mtodos numricos**: La mayora de las tcnicas para resolver numricamente a las [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) se basan en la idea
de discretizar el problema en cada variable independiente que se produce en la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales), y de esta forma, reformular el problema en una forma algebraica. Esto usualmente resulta en problemas de [lgebra lineal](https://es.wikipedia.org/wiki/%C3%81lgebra_lineal) de gran escala. Dos  de las tcnicas principales para reformular las [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) a una forma algebraica son, los [mtodos de diferencias finitas](https://es.wikipedia.org/wiki/M%C3%A9todo_de_las_diferencias_finitas) (MDF), donde las [derivadas](https://es.wikipedia.org/wiki/Derivada) del problema son aproximadas por medio de la frmula de [diferencias finitas](https://es.wikipedia.org/wiki/Diferencia_finita); y los [mtodos de los elementos finitos](https://es.wikipedia.org/wiki/M%C3%A9todo_de_los_elementos_finitos) (MEF), en donde la funcin incgnita se escribe como combinacin lineal de funciones de base simple que pueden ser derivadas e integradas fcilmente. En muchos casos, estos mtodos numricos van a ser las nicas herramientas que podamos utilizar para resolver a las [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales).

### Solucin de ecuaciones en derivadas parciales bsicas 

Como resolver este tipo de ecuaciones es una tarea realmente complicada, vamos a empezar por resolver analticamente las ms fciles para ganar confianza y as luego poder pasar a ecuaciones ms complicadas. 

La ms simple de las [Ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) que nos podemos encontrar es:

$u_x = 0$, en donde $u = u(x, y)$.

Esta ecuacin nos dice que la [derivada parcial](https://es.wikipedia.org/wiki/Derivada_parcial) de $u$ con respecto a $x$ es cero, lo que significa que $u$ no depende de $x$. Por lo tanto, la solucin de esta ecuacin va a ser $u=f(y)$, en donde $f$ es una funcin arbitraria de una variable. Por ejemplo, $u = y^2 - y$ podra ser una posible solucin.

Subiendo un poco ms la complejidad, podemos pasar a una [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de segundo orden, como la siguiente:

$u_{xx} = 0$

En este caso, podemos integrar una vez $u_{xx}$ para obtener $u_x(x, y) = f(y)$. Si volvemos a integrar este resultado, podemos arribar a la solucin final $u(x, y) = f(y)x + g(y)$, en donde $f$ y $g$ son dos funciones arbitrarias.

Por ltimo, si quisiramos resolver la siguiente [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales):

$u_{xy} = 0$

Primero integramos con respecto a $x$ tomando a $y$ como fija, de esta forma obtenemos $u_y(x, y) = f(y)$. Luego podemos integrar con respecto a $y$ tomando a $x$ como fija, y llegamos a la solucin:

$u(x, y) = F(y) + g(x)$, en donde $F' = f$

Como podemos ver de estos ejemplos, las soluciones analticas de las [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) dependen de **funciones arbitrarias** (en lugar de *<a href="https://es.wikipedia.org/wiki/Constante_(matem%C3%A1ticas)">constantes</a>* arbitrarias como era el caso de las [EDO](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/)). Por lo tanto vamos a necesitar condiciones auxiliares para poder determinar una nica solucin.

### La condicin inicial y la condicin de frontera

Al igual que nos pasaba cuando vimos las [ecuaciones diferenciales ordinarias](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/); las [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) pueden tener muchas soluciones, pero a nosotros nos va a interesar encontrar la solucin para un caso particular; para lograr esto, debemos imponer unas condiciones auxiliares al problema original. Estas condiciones van a estar motivadas por la [Fsica](https://es.wikipedia.org/wiki/F%C3%ADsica) del problema que estemos analizando y pueden llevar a ser de dos tipos diferentes: [condiciones iniciales](https://en.wikipedia.org/wiki/Initial_condition) y [condiciones de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera).

La [condicin inicial](https://en.wikipedia.org/wiki/Initial_condition) va a establecer el estado del problema al momento de tiempo cero, $t_0$. Por ejemplo para el problema de *[difusin](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_de_difusi%C3%B3n)*, la [condicin inicial](https://en.wikipedia.org/wiki/Initial_condition) va a ser:

$$u(x, t_0) = \phi(x)$$

donde $\phi(x)= \phi(x, y, z)$ es una funcin que puede representar el estado de concentracin inicial. Para el problema del *[flujo del calor](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_del_calor)*, $\phi(x)$ va a representar la temperatura inicial.



La [condicin de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera) nos va a delimitar el *dominio* en el que nuestra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) es vlida. As por ejemplo, volviendo al problema de *[difusin](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_de_difusi%C3%B3n)*, el *dominio* en el que nuestra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) es vlida, puede estar delimitado por la superficie del objeto que contiene al lquido. Existen varios tipos de [condiciones de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera), de las cuales las ms importantes son:

* La **[condicin de frontera de Dirichlet](https://es.wikipedia.org/wiki/Condici%C3%B3n_de_frontera_de_Dirichlet)**, en dnde los valores vlidos de la funcin incgnita $u$ son especificados.

* La **[condicin de frontera de Neumann](https://es.wikipedia.org/wiki/Condici%C3%B3n_de_frontera_de_Neumann)**, en donde los valores vlidos especificados son dados para alguna de las derivadas de $u$.

* La **[condicin de frontera de Robin](https://es.wikipedia.org/wiki/Condici%C3%B3n_de_frontera_de_Robin)**, en donde los valores vlidos son especificados por una [combinacin lineal](https://es.wikipedia.org/wiki/Combinaci%C3%B3n_lineal) de una funcin y las derivadas de $u$.

### Interpretacin geomtrica de EDP de primer orden 

Las [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de primer orden poseen una interpretacin geomtrica la cual nos puede facilitar alcanzar una solucin general para ellas. 

#### Coeficientes constantes

Tomemos la siguiente ecuacin de coeficientes constantes:

$au_x + bu_y = 0$, en donde $a$ y $b$ son *<a href="https://es.wikipedia.org/wiki/Constante_(matem%C3%A1ticas)">constantes</a>* y ambas no pueden ser cero.

En esta ecuacin la cantidad  $au_x + bu_y$ es la derivada direccional de $u$ en la direccin del [vector](https://es.wikipedia.org/wiki/Vector) $V = (a, b) = ai + bj$. Como esta cantidad tiene que ser cero, esto significa que $u (x, y)$ debe ser constante en la direccin de $V$. El [vector](https://es.wikipedia.org/wiki/Vector) $(b, -a)$ es <a href="https://es.wikipedia.org/wiki/Ortogonalidad_(matem%C3%A1ticas)">ortogonal</a> a $V$, por lo tanto, las lneas paralelas a $V$ (ver el grfico ms abajo) tienen las ecuaciones $bx - ay$ constantes. (Se las llama las *lneas caractersticas*.) Entonces, la solucin es constante en cada una de esas lneas. Por lo tanto, $u (x, y)$ depende de $bx - ay$ solamente. De esta forma podemos llegar a la solucin general para este tipo de ecuaciones, que va a ser:

$u(x, y) = f(bx - ay)$, en donde $f$ es una funcin arbitraria de una variable.


<img alt="EDP de 1er orden con coef constantes" title="EDP de 1er orden con coef constantes" src="http://relopezbriega.github.io/images/PDE_Const.png
" >

Entonces, si por ejemplo, quisiramos resolver la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales), $4u_x - 3u_y= 0$, con la [condicin de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera) que $u(0, y) = y^3$. Podemos aplicar la solucin general que obtuvimos arriba y llegar al resultado $u(x, y) = f(-3x - 4y)$. Ahora, solo nos faltara aplicar la condicin para poder determinar cual es la funcin arbitraria $f$. Si sustituimos $x=0$ en nuestra solucin, obtenemos $y^3 = f(-4y)$. Si decimos que $z = -4y$, entonces nuestra funcin es $f(z) = -z^3 / 64$. Por lo tanto la solucin de nuestra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) que satisface la [condicin de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera) es $u(x, y) = (3x + 4y)^3 / 64$. 

#### Coeficientes variables

Consideremos ahora la siguiente [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) de coeficientes variables:

$u_x + yu_y = 0$

Esta ecuacin es similar a la que vimos anteriormente, con la diferencia de que ahora tenemos al coeficiente variable $y$. Utilizando la misma intuicin geomtrica que usamos antes, podemos ver que aqu tambin la derivada direccional de $u$ en el vector $v=(1, y)$ es constante, pero esta vez el vector no es constante, sino que es variable con $y$. Las curvas que tienen a $v$ como su vector tangente, tienen pendiente $y/1$, es decir:

$\frac{dy}{dx} = \frac{y}{1}$.

Podemos resolver esta ecuacin como una [EDO](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_diferencial_ordinaria) y as obtener:

$y = Ce^x$, o lo que es lo mismo $e^{-x}y = C$.

La solucin de nuestra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) entonces va a ser constante en estas *curvas caractersticas* (ver grfico); y van a responder a la siguiente solucin general:

$u(x, y) = f(e^{-x}y)$, en donde $f$ es una funcin arbitraria.

### Separacin de variables

Otra forma en que nos podemos ayudar de [SymPy](http://www.sympy.org/es/) para resolver [EDPs](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales), es utilizando el [mtodo de separacin de variables](https://es.wikipedia.org/wiki/M%C3%A9todo_de_separaci%C3%B3n_de_variables). La caracterstica esencial de esta tcnica es transformar la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) en un conjunto de [EDOs](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/) (las cuales podemos solucionar con la ayuda de [SymPy](http://www.sympy.org/es/)). De esta forma, la solucin requerida de la [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) se expresa como
un producto $u (x, y) = X (x) Y (y) \ne 0$, o como una suma $u (x, y) = X (x) + Y (y)$, donde $X (x)$ e $Y (y)$ son funciones de $x$ e $y$, respectivamente. Muchos de los problemas significativos en [ecuaciones en derivadas parciales](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) pueden ser resueltos por este mtodo. Para ilustrar como funciona esta tcnica, veamos un ejemplo. Vamos a resolver la siguiente [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales).

$y^2u_x^2 + x^2u_y^2 = (xyu)^2$, que cumple con la condicin $u(x, 0) = 3e^{x^2/4}$.

Podemos entonces asumir que $u(x, y) = f(x) g(y) \ne 0$ es una *solucin separable* de ella; por tanto lo reemplazamos en la ecuacin para obtener:

$$y^2(f'(x) g(y))^2 + x^2(f(x) g'(y))^2 = x^2 y^2 (f(x) g(y))^2$$ 

lo que es equivalente a decir;

$$\frac{1}{x^2}\left(\frac{f'(x)}{f(x)}\right)^2 + \frac{1}{y^2}\left(\frac{g'(y)}{g(y)}\right)^2 = 1$$

Luego, ayudndonos de la *constante de separacin* $\lambda^2$, podemos separar a esta ecuacin en dos [EDOs](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/), del siguiente modo; primero igualamos la ecuacin anterior a $\lambda^2$:

$$\frac{1}{x^2}\left(\frac{f'(x)}{f(x)}\right)^2 = 1 - \frac{1}{y^2}\left(\frac{g'(y)}{g(y)}\right)^2 = \lambda^2$$

y luego separamos ambas ecuaciones para obtener:

$$\frac{1}{x}\frac{f'(x)}{f(x)} = \lambda \\ \frac{g'(x)}{yg(y)}= \sqrt{1 - \lambda^2}$$

Ahora podemos utilizar `sympy.dsolve` para resolver ambas [EDOs](http://relopezbriega.github.io/blog/2016/01/10/ecuaciones-diferenciales-con-python/):

## Ejemplos

````python
# EDO n 1
edo1 = sympy.Eq((1 / x) * (f(x).diff(x)/f(x)) - z)
edo1
````

````python
# Resolviendo EDO n 1
sympy.dsolve(edo1)
````

````python
# EDO n 2
edo2 = sympy.Eq((f(y).diff(y)) / (y*f(y)) - sympy.sqrt(1 - z**2))
edo2
````

````python
# Resolviendo EDO n 2
sympy.dsolve(edo2)
````

Entonces ahora podemos utilizar estos resultados para armar la solucin final a nuestra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) original, el cual va a ser:

$$u(x, y) = C e^{\frac{\lambda}{2}x^2 + \frac{1}{2}y^2\sqrt{1 - \lambda^2}}$$

En donde $C = C_1 C_2$ es una *constante arbitraria*.

Por ltimo, utilizando la condicin $u(x, 0) = 3e^{x^2/4}$, podemos despejar tanto a $C$ ($C=3$) como a $\lambda$ ($\lambda = 1/2$) y arribar a la solucin final:

$$u(x, y) = 3 e^{\frac{1}{4}\left(x^2 + y^2\sqrt{3}\right)}$$

Si bien debemos realizar un trabajo manual previo, aun as [SymPy](http://www.sympy.org/es/) sigue siendo de gran ayuda para facilitarnos llegar a la solucin final.

### Mtodos numricos - Mtodo de Elementos Finitos

Por ltimo, para cerrar este artculo, veamos como podemos aplicar el [mtodos de los elementos finitos](https://es.wikipedia.org/wiki/M%C3%A9todo_de_los_elementos_finitos) (MEF) con [Python](https://www.python.org/). Para esto nos vamos ayudar de la librera [FEniCS](http://fenicsproject.org/), la cual es un *framework* para resolver numricamente problemas generales de [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) utilizando el [mtodos de los elementos finitos](https://es.wikipedia.org/wiki/M%C3%A9todo_de_los_elementos_finitos). Deben tener en cuenta que por ahora solo funciona con [Python 2](https://www.python.org/downloads/release/python-2711/). La interfaz principal que vamos a utilizar para trabajar con este *framework* nos la proporcionan las libreras `dolfin` y `mshr`; las cuales debemos importar para poder trabajar con el. Una vez importadas, podemos configurar algunos de sus parmetros para lograr el comportamiento deseado.

````python
# importando modulos de fenics
import dolfin
import mshr

dolfin.parameters["reorder_dofs_serial"] = False
dolfin.parameters["allow_extrapolation"] = True
````

El problema que vamos a resolver con la ayuda de [FEniCS](http://fenicsproject.org/), va a ser la siguiente [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales):

$$u_{xx} + u_{yy} = 0$$

Con las siguientes [condiciones de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera):

$$u(x=0) = 3 ; \ u(x=1)=-1 ; \ u(y=0) = -5 ; \ u(y=1) = 5$$  

El primer paso en la solucin de una [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) utilizando el [mtodos de los elementos finitos](https://es.wikipedia.org/wiki/M%C3%A9todo_de_los_elementos_finitos),  es definir una *[malla](https://es.wikipedia.org/wiki/Malla_poligonal)* que describa la discretizacin del *dominio* del problema. Para este caso, vamos a utilizar la funcin `RectangleMesh` que nos ofrece [FEniCS](http://fenicsproject.org/).

````python
# Discretizando el problema
N1 = N2 = 75
mesh = dolfin.RectangleMesh(dolfin.Point(0, 0), dolfin.Point(1, 1), N1, N2)
````

El siguiente paso es definir una representacin del [espacio funcional](https://es.wikipedia.org/wiki/Espacio_funcional) para las funciones de ensayo y prueba. Para esto vamos a utilizar la clase `FunctionSpace`. El constructor de esta clase tiene al menos tres argumentos: un objeto de *[malla](https://es.wikipedia.org/wiki/Malla_poligonal)*, el nombre del tipo de funcin base, y el grado de la funcin base. En este caso, vamos a utilizar la funcin de [Lagrange](https://es.wikipedia.org/wiki/Interpolaci%C3%B3n_polin%C3%B3mica_de_Lagrange).

````python
# Funciones bases
V = dolfin.FunctionSpace(mesh, 'Lagrange', 1)
u = dolfin.TrialFunction(V)
v = dolfin.TestFunction(V)
````

Ahora debemos definir a nuestra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) en su [formulacin dbil](https://es.wikipedia.org/wiki/Formulaci%C3%B3n_d%C3%A9bil_de_una_ecuaci%C3%B3n_diferencial) equivalente para poder tratarla como un problema de [lgebra lineal](https://es.wikipedia.org/wiki/%C3%81lgebra_lineal) que podamos resolver con el [MEF](https://es.wikipedia.org/wiki/M%C3%A9todo_de_los_elementos_finitos).

````python
# Formulacin debil de la EDP
a = dolfin.inner(dolfin.nabla_grad(u), dolfin.nabla_grad(v)) * dolfin.dx
f = dolfin.Constant(0.0)
L = f * v * dolfin.dx
````

Por ltimo, solo nos falta definir las [condiciones de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera).

````python
# Defino condiciones de frontera
def u0_top_boundary(x, on_boundary):
    return on_boundary and abs(x[1]-1) < 1e-8

def u0_bottom_boundary(x, on_boundary):
    return on_boundary and abs(x[1]) < 1e-8

def u0_left_boundary(x, on_boundary):
    return on_boundary and abs(x[0]) < 1e-8

def u0_right_boundary(x, on_boundary):
    return on_boundary and abs(x[0]-1) < 1e-8
````

````python
# Definiendo condiciones de frontera de Dirichlet
bc_t = dolfin.DirichletBC(V, dolfin.Constant(5), u0_top_boundary)
bc_b = dolfin.DirichletBC(V, dolfin.Constant(-5), u0_bottom_boundary)
bc_l = dolfin.DirichletBC(V, dolfin.Constant(3), u0_left_boundary)
bc_r = dolfin.DirichletBC(V, dolfin.Constant(-1), u0_right_boundary)

# Lista de condiciones de frontera
bcs = [bc_t, bc_b, bc_r, bc_l]
````

Con esta especificacin de las [condiciones de frontera](https://es.wikipedia.org/wiki/Problema_de_condici%C3%B3n_de_frontera), ya estamos listos para resolver nuestra [EDP](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n_en_derivadas_parciales) utilizando la funcin `dolfin.solve`. El vector resultante, luego lo podemos convertir a una matriz de [NumPy](http://www.numpy.org/) y utilizarla para graficar la solucin con [Matplotlib](http://matplotlib.org/).

````python
# Resolviendo la EDP
u_sol = dolfin.Function(V)
dolfin.solve(a == L, u_sol, bcs)
````

````python
# graficando la solucin
u_mat = u_sol.vector().array().reshape(N1+1, N2+1)

x = np.linspace(0, 1, N1+2)
y = np.linspace(0, 1, N1+2)
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots(1, 1, figsize=(8, 6))

c = ax.pcolor(X, Y, u_mat, vmin=-5, vmax=5, cmap=mpl.cm.get_cmap('RdBu_r'))
cb = plt.colorbar(c, ax=ax)
ax.set_xlabel(r"$x_1$", fontsize=18)
ax.set_ylabel(r"$x_2$", fontsize=18)
cb.set_label(r"$u(x_1, x_2)$", fontsize=18)
fig.tight_layout()
````
